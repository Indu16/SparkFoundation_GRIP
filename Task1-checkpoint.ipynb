# Task1- Prediction Using Supervised Machine Learning

Import the required libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

Read data

```python
df=pd.read_csv('http://bit.ly/w-data')
df.head()
```

Check if there is a NULL value in the dataset

```python
df.isnull== True
```

There are no NULL values in the dataset

Visualize the data

```python
sns.set_style('whitegrid')
sns. scatterplot(x = "Hours", y = "Scores", data = df)
plt.title('Graph : Hours vs Percentage',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

From the above scatterplot,there seems to be a corelation between 'Percentage of Marks' and 'Study Hours'.

Now , we plot a Regression Line to confirm the corelation. 

```python
sns.regplot(x= 'Hours', y= 'Scores', data= df)
plt.title('Regression Plot',size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
print(df.corr())
```

Now its confirmed that the variables are positively corelated.

TRAINING THE MODEL

Splitting the data

```python
# Defining X and y from the Data
X = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values

# Spliting the Data in two
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

Fit data into the model

```python
regression = LinearRegression()
regression.fit(train_X, train_y)
print("---------Model Trained---------")
```

Predicting marks

```python
pred_y = regression.predict(val_X)
prediction = pd.DataFrame({'Hours': [i[0] for i in val_X], 'Predicted Marks': [k for k in pred_y]})
prediction
```

Compare predicted and actual marks

```python
compare_scores = pd.DataFrame({'Actual Marks': val_y, 'Predicted Marks': pred_y})
compare_scores
```

Comparing through Graphs

```python
plt.scatter(x=val_X, y=val_y, color='blue')
plt.plot(val_X, pred_y, color='Black')
plt.title('Actual vs Predicted', size=20)
plt.ylabel('Marks Percentage', size=12)
plt.xlabel('Hours Studied', size=12)
plt.show()
```

Evaluating the model

```python
#Calculating the accuracy of the model
print('Mean absolute error: ',mean_absolute_error(val_y,pred_y))
```

### # What will be the predicted score of a student if he/she studies for 9.25 hrs/ day?

```python
hours = [9.25]
answer = regression.predict([hours])
print("Score = {}".format(round(answer[0],3)))
```

According to the Regression Model, if a student studies for 9.25 hours a day, he/she is likely to score 93.893 marks.
